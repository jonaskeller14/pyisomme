```python
import pyisomme
import copy
import matplotlib.pyplot as plt
```

```python
v11391 = pyisomme.Isomme().read("../data/nhtsa/11391/11391.mme")
```

## Raw Channel

```python
channel = v11391.get_channel("11HEAD*")
```

## CFC 1000 / Filterclass A

```python
channel_cfc1000 = copy.deepcopy(channel).cfc(1000)
```

## CFC 60 / Filterclass D

```python
channel_cfc60 = copy.deepcopy(channel).cfc("D")
```

## Plot

```python
fig, axes = plt.subplots(nrows=3,ncols=1, figsize=(12,12))

channel.data.plot(ax=axes[0], subplots=True, color="r")

# channel.data.plot(ax=axes[1], subplots=True,color="r")
channel_cfc1000.data.plot(ax=axes[1], subplots=True)

# channel.data.plot(ax=axes[2], subplots=True,color="r")
channel_cfc60.data.plot(ax=axes[2], subplots=True)

plt.show()
```
