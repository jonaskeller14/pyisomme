```python
import pyisomme
```

```python
channel_1 = pyisomme.create_sample("1???????????????", t_range=(0,0.1,1000), y_range=(0,10), mode="linear", unit="m/(s*s)")
channel_1.data.plot()

channel_2 = pyisomme.create_sample("2???????????????", t_range=(0,0.1,700), y_range=(-10,10), mode="sin", unit="m/(s*s)")
channel_2.data.plot()
```

# Addition

```python
channel_res = channel_1 + channel_2 + 1
channel_res.data.plot()
```

# Subtraction

```python
channel_res = channel_1 - channel_2 - 1
channel_res.data.plot()
```

# Multiplication

```python
channel_res = channel_1 * channel_2 * 2
channel_res.data.plot()
```

# Division

```python
channel_res = channel_1 / channel_2 / 2
channel_res.data.plot()
```

# Exponent

```python
channel_res = channel_1**2
channel_res.data.plot()
```
