```python
import pyisomme
```

# Read

```python
v1 = pyisomme.Isomme().read("../data/nhtsa/11391/11391.mme", "11NECKUP????FO??", "11TIBI*FO*")
v2 = pyisomme.Isomme().read("../data/nhtsa/14084/14084.mme", "11NECKUP????FO??", "11TIBI*FO*")
v3 = pyisomme.Isomme().read("../data/nhtsa/14065/14065.mme", "11NECKUP????FO??", "11TIBI*FO*")

v1.print_channel_list()
v2.print_channel_list()
v3.print_channel_list()
```

# Plot

```python
fig = pyisomme.Plot_Line({isomme: [[isomme.get_channel("11NECKUP????FOX?")]] for isomme in (v1, v2, v3)})
```

```python
fig = pyisomme.Plot_Line({isomme: [[isomme.get_channel(f"11NECKUP0000FO{xyzr}D")] for xyzr in "XYZR"] for isomme in (v1, v2, v3)}, 2, 2)  # Plot filtered channels, 2x2-Layout
```

```python
fig = pyisomme.Plot_Line({isomme: [[f"11NECKUP0000FO{xyzr}D"] for xyzr in "XYZR"] for isomme in (v1, v2, v3)}, 4, 1)  # Plot filtered channels, 4x1-Layout
```

```python
fig = pyisomme.Plot_Line({isomme: [
    [isomme.get_channel("11TIBILEUP??FOZD")], 
    [isomme.get_channel("11TIBIRIUP??FOZD")], 
    [isomme.get_channel("11TIBILELO??FOZD")], 
    [isomme.get_channel("11TIBIRILO??FOZD")]] for isomme in (v1, v2, v3)}, 2, 2,  xlim=(0, 20), sharey=True)
```

```python
fig = pyisomme.Plot_Line({isomme: [["11TIBILEUP??FOZD", "11TIBIRIUP??FOZD", "11TIBILELO??FOZD", "11TIBIRILO??FOZD"]] for isomme in (v1, v2, v3)},  xlim=(0, 160), sharey=True)
```
