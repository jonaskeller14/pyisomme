```python
import pyisomme
```

# Read

```python
v1 = pyisomme.Isomme().read("../data/nhtsa/11391", "1*")
v2 = pyisomme.Isomme().read("../data/nhtsa/14084", "1*")
```

# Report

```python
report = pyisomme.report.euro_ncap.frontal_50kmh.EuroNCAP_Frontal_50kmh([v1, v2])

# Set manual inputs (info which can not be read from curve data - optional)
report.criterion_master[v1].criterion_door_opening_during_impact.number_of_door_openings_during_impact = 1
report.criterion_master[v2].criterion_driver.criterion_femur.criterion_submarining.submarining = True
```

```python
report.calculate()
```

```python
report.print_results()
```
